## 1. 경계선을 구분하여 이미지 검출 
## Canny Edge Dection

```python
import cv2

img = cv2.imread('snowman.jpg')

#대상이미지, minVal하위임계값(이하면 경계선으로 생각,검출), maxVal상위임계값(이상이면 경계선으로 생각)
canny = cv2.Canny(img, 150, 200)
    
cv2.imshow('img',img)
cv2.imshow('canny',canny)
    
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

def empty(pos):
    pass

img = cv2.imread('snowman.jpg')

name = 'tracker'
cv2.namedWindow(name)

cv2.createTrackbar('threshold1',name,0,255,empty) #minVal
cv2.createTrackbar('threshold2',name,0,255,empty) #maxVal

while True:
    threshold1 = cv2.getTrackbarPos('threshold1',name)
    threshold2 = cv2.getTrackbarPos('threshold2',name)
    
    canny = cv2.Canny(img, threshold1, threshold2)
    
    cv2.imshow('img',img)
    cv2.imshow(name,canny)
    
    if(cv2.waitKey(1)==ord('q')):
        break
        
cv2.destroyAllWindows()
```

## 2. 윤곽선(contour) 경계선을 연결한 선 
## 을 통해 이미지 검출


```python
import cv2

img = cv2.imread('card.png')
target_img = img.copy() #사본이미지

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #gray로

#오츠알고리즘으로 이진화 깔끔하게
ret,otsu = cv2.threshold(gray,-1,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)

#윤곽선 검출
#윤곽선 정도, 계층구조
#이미지,윤곽선찾기모드,윤곽선찾을때사용하는근사치방법
counturs,hierarchy = cv2.findContours(otsu,cv2.RETR_LIST,cv2.CHAIN_APPROX_NONE)

color = (0,255,0) #green
#윤곽선 그리기
#대상이미지, 윤곽선정보, 인덱스(-1아면 전체), 색깔, 두께
cv2.drawContours(target_img, counturs, -1, color,2)


cv2.imshow('img',img)
cv2.imshow('gray',gray)
cv2.imshow('otsu',otsu)
cv2.imshow('counturs',target_img)


cv2.waitKey(0)
cv2.destroyAllWindows()

```

### 윤곽선찾기모드
1. cv2.RETR_EXTERNAL 가장외곽의 윤곽선만 찾음
2. cv2.RETR_LIST 모든 윤곽선을 찾음 계층정보없음
3. cv2.RETR_TREE 모든 윤곽선을 찾음 계층정보를 트리구조로 생성

```python
import cv2

img = cv2.imread('card.png')
target_img = img.copy() #사본이미지

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #gray로

#오츠알고리즘으로 이진화 깔끔하게
ret,otsu = cv2.threshold(gray,-1,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)

#윤곽선 검출
#윤곽선 정도, 계층구조
#이미지,윤곽선찾기모드,윤곽선찾을때사용하는근사치방법
counturs,hierarchy = cv2.findContours(otsu,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)

color = (0,255,0) #green
#윤곽선 그리기
#대상이미지, 윤곽선정보, 인덱스(-1아면 전체), 색깔, 두께
cv2.drawContours(target_img, counturs, -1, color,2)


cv2.imshow('img',img)
cv2.imshow('gray',gray)
cv2.imshow('otsu',otsu)
cv2.imshow('counturs',target_img)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 경계 사각형 (윤곽선의 경계면을 둘러싼 사각형) 그리기
boundingRect()

```python
import cv2

img = cv2.imread('card.png')
target_img = img.copy() #사본이미지

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #gray로
ret,otsu = cv2.threshold(gray,-1,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)
contours,hierarchy = cv2.findContours(otsu,cv2.RETR_LIST,cv2.CHAIN_APPROX_NONE)

color = (0,255,0) #green
#윤곽선 그리기
#대상이미지, 윤곽선정보, 인덱스(-1아면 전체), 색깔, 두께
#cv2.drawContours(target_img, contours, -1, color,2)

for cnt in contours :
    x,y,width,height = cv2.boundingRect(cnt)
    #사각형 그림
    cv2.rectangle(target_img,(x,y),(x+width,y+height),color,2)

cv2.imshow('contours',target_img)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 면적 크기
contourArea()

```python
import cv2

img = cv2.imread('card.png')
target_img = img.copy() #사본이미지

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #gray로
ret,otsu = cv2.threshold(gray,-1,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)
contours,hierarchy = cv2.findContours(otsu,cv2.RETR_LIST,cv2.CHAIN_APPROX_NONE)

color = (0,255,0) #green

for cnt in contours :
    # 면적 2500이상인 경계사각형만 찾기
    if cv2.contourArea(cnt)>25000:
        x,y,width,height = cv2.boundingRect(cnt) 
        #사각형 그림
        cv2.rectangle(target_img,(x,y),(x+width,y+height),color,2)

cv2.imshow('contours',target_img)


cv2.waitKey(0)
cv2.destroyAllWindows()
```
