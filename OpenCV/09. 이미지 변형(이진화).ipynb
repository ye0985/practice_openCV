```python
#trackbar를 이용해 실시간으로 임계값바꾸기
import cv2

def empty(pos):
    print(pos)
    pass

img = cv2.imread('book.jpg',cv2.IMREAD_GRAYSCALE) #그레이로 가져오기

name = 'trackbar'
cv2.namedWindow(name) #윈도우창 만들기

#trackbar이름, 띄울 창위치, 시작값, 최대값, 이벤트처리
cv2.createTrackbar('threshold', name, 127, 255, empty) #trackbar 만들기

while True: 
    #실시간으로 창에서 trackbar 받아와 저장
    thresh = cv2.getTrackbarPos('threshold',name) #trackbar이름, 창이름
    # 저장한 것을 binary값으로
    ret, binary = cv2.threshold(img, thresh, 255, cv2.THRESH_BINARY) #임계값은 저장값으로
    
    if not ret:
        break # 반복문 끝
        
    cv2.imshow(name,binary) #윈도우창에 binary보여주기
    
    if cv2.waitKey(1) == ord('q'):
        break

cv2.DestroyAllWindows()
```

## Adaptive Threshold 글자가 잘 보이도록
이미지를 작은 영역으로 나누어 임계치 적용

```python
import cv2

def empty(pos):
    #print(pos)
    pass

img = cv2.imread('book.jpg',cv2.IMREAD_GRAYSCALE) #그레이로 가져오기

name = 'trackbar'
cv2.namedWindow(name) #윈도우창 만들기

#trackbar이름, 띄울 창위치, 시작값, 최대값, 이벤트처리
cv2.createTrackbar('block_size', name, 25, 255, empty) #시작값: 홀수만 가능,1보다 큰값
cv2.createTrackbar('c', name, 3, 10, empty) #시작값: 일반적으로 양수 사용


while True: 
    #실시간으로 창에서 trackbar 받아와 저장
    block_size = cv2.getTrackbarPos('block_size',name) #trackbar이름, 창이름
    c = cv2.getTrackbarPos('c',name)
    
    if block_size <= 1: #1이하면 3으로
        block_size = 3

    if block_size % 2 == 0: #짝수면 홀수로
        block_size += 1
        
    # 저장한 것을 binary값으로 Adaptive Threshold
    # 이미지, 임계값이상일때 색깔(흰색), cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, c
    binary = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, c)
       
    cv2.imshow(name,binary) #윈도우창에 binary보여주기
    
    if cv2.waitKey(1) == ord('q'):
        break

cv2.DestroyAllWindows()
```

## 오츠 알고리즘 : 적절한 임계값을 자동 적용

```python
import cv2
img = cv2.imread('book.jpg',cv2.IMREAD_GRAYSCALE) #그레이로 가져오가

#저장위치, 임계값(흑백 나눌 기준점), 최대값(0-255중), binary화
ret, binary = cv2.threshold(img, 127,255,cv2.THRESH_BINARY)
# 오츠알고리즘 사용
ret, otsu = cv2.threshold(img, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU) #임계값자리는 무시됨
print('otsu threshold', ret)

cv2.imshow('img',img)
cv2.imshow('binary',binary)
cv2.imshow('otsu',otsu)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
